— The essence of gradient descent algorithm —

The essence of the gradient descent algorithm is to iteratively adjust parameters in the direction opposite to the gradient of a cost function, aiming to find the minimum of the function. It is well-rooted in many advanced machine learning algorithms such as gradient boosting trees, deep neural networks, etc. Here, you can find a simple example illustrating the essence mechanism behind the algorithm. 

```python
# Gradient descent algorithm 
import numpy as np 

def f(x):
    return (x-5)**2 + 1

def gradient(x):
    return 2 * (x-5) # derivative of (x-5)**2 + 1
    
x = np.random.uniform(low=-5, high=15) #initial x
epochs = 100 # number of iterations 
lr = 0.1 # learning rate

x_values = []
y_values = []

for epoch in range(epochs):
    g = gradient(x)  # compute gradient at x
    x = x - lr * g # update x 
    x_values.append(x)
    y_values.append(f(x))

print('Gradient descent results for minimizing f(x) = (x-5)**2 + 1}:')
print(f'x = {x:.2f}, f(x): {f(x):.2f}')
```

```python
# Animation
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation


fig, ax = plt.subplots()
line, = ax.plot([], [], 'ro-', label='Optimization', lw = 2)
ax.plot(np.linspace(-5, 15, 100), f(np.linspace(-5, 15, 100)), label='Objective Function: $%s$' %r'(x-5)^2 + 1', lw = 2)

def init():
    ax.set_xlim(-6, 16)
    ax.set_ylim( 0 , f(16) + 4)
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.set_title('Gradient Descent Progression')
    ax.legend()
    return line,

def update(frame):
    line.set_data(x_values[:frame], y_values[:frame])
    return line,

animation = FuncAnimation(fig, update, frames=len(x_values), init_func=init, blit=True, interval=10)

animation.save('gradient_descent_progression.mp4', writer='ffmpeg', fps=5 , dpi = 300)

plt.show()
```
